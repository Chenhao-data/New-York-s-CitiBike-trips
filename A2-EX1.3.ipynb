ax2.plot(df_temp["avgTemp"], color="red"): This line plots the average temperature data in red on the secondary y-axis.

# Bonus Task

## Create a histogram of the variable “tripduration” 

```python
# 1. Convert 'started_at' and 'ended_at' to datetime format
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# 2. Create 'tripduration' column as the difference in seconds
df['tripduration'] = (df['ended_at'] - df['started_at']).dt.total_seconds()
```

```python
plt.figure(figsize=(8, 5))  # Set the figure size
plt.hist(df['tripduration'], bins=30, color='skyblue', edgecolor='black')  # Create histogram
plt.title('Distribution of Trip Duration', fontsize=16)  # Add a title
plt.xlabel('Trip Duration (in seconds)', fontsize=14)  # Label for x-axis
plt.ylabel('Frequency', fontsize=14)  # Label for y-axis
plt.show()
```

## Add a curve following the bins

```python
import seaborn as sns
```

```python
# 1. Plot the histogram
plt.hist(df['tripduration'], bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# 2. Add a KDE curve using seaborn
sns.kdeplot(df['tripduration'], color='red', linewidth=2)

# Add labels and title
plt.title('Trip Duration with KDE Curve', fontsize=16)
plt.xlabel('Trip Duration (in seconds)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Show plot
plt.show()
```
